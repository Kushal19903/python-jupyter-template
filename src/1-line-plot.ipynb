### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import numpy as np
import matplotlib.pyplot as plt


```

```python
# Sample Data
x = np.arange(0, 11)
y = x ** 2
```

```python

# Error Handling
if not isinstance(x, (list, np.ndarray)) or not isinstance(y, (list, np.ndarray)):
    raise TypeError("x and y must be lists or numpy arrays")
if len(x) != len(y):
    raise ValueError("x and y must have the same length")
```

```python

# Plot
plt.plot(x, y, label="y = x^2")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```
