### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np


```

```python

# Generate x values from 0 to 10
x = np.arange(0, 11, 1)
y = x ** 2  # Square of x
```

```python

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')
```

```python

# Customize the plot
plt.title("Line Plot: y = x²")  # Title of the plot
plt.xlabel("X-axis (Input values)")  # Label for X-axis
plt.ylabel("Y-axis (Squared values)")  # Label for Y-axis
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid with dashed lines
plt.legend()  # Show legend

# Display the plot
plt.show()
```
