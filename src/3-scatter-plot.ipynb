### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

```python
# Error Handling
if len(x) != len(y):
    raise ValueError("x and y must have the same length")
if not all(isinstance(i, (int, float)) for i in x + y):
    raise TypeError("x and y values must be numeric")
```

```python

# Plot
plt.scatter(x, y, color='blue')
for i, txt in enumerate(y):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0,5), ha='center')

plt.title("Scatter Plot with Annotations")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
