### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt
```

```python

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
```

```python

# Error Handling
if not all(isinstance(label, str) for label in labels):
    raise TypeError("Labels must be strings")
if not all(isinstance(size, (int, float)) for size in sizes):
    raise TypeError("Sizes must be integers or floats")
if len(labels) != len(sizes):
    raise ValueError("Labels and sizes must have the same length")
```

```python

# Plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'purple'])
plt.title("Pie Chart")
plt.show()
```
